## MATRICES

### Introduction to matrices

#### Ma trận là gì?

Ma trận là một mảng chữ nhật gồm các phần tử đuọc sắp xếp theo hàng và cột. Phần từ có thể là số, ký tự, expression. Dưới đây là phần tử số (numerical element)


<img src="img/intro2matrix.svg" alt="drawing" width="400"/>


$
A_{4,3} = \begin{pmatrix}
        8 & 21  & 9          \\[0.3em]
        5 & 2  & 11         \\[0.3em]
        12 &  4 & 11        \\[0.3em]
        2 & 15 & 4
     \end{pmatrix}$

#### Chiều của ma trận

Số dòng và số cột được gọi là chiều của ma trận.

#### Vectors

Ma trận mà chỉ có một dòng/ một cột thì được gọi là Vectors (row vector/ column vector).

Ví dụ: điểm $K \begin{pmatrix}7 & 0 & -13 \end{pmatrix}$ được gọi là "one by three" matrix.

#### Matrix Elements

Phần tử có một ví trí của riêng nó trong ma trận. Trong ma trận A, phần tử thuộc dòng m cột n được kí hiệu là $a_{mn}$. 

Ví dụ $a_{12} = 21$ trong ma trận $A$

### Type of matrices

#### Basic matrices

- **row matrix** là ma trận chỉ có một hàng

$\begin{pmatrix} 1 & 0 & 3 \end{pmatrix}$

- **column matrix** là ma trận chỉ có một cột

$\begin{pmatrix} 1 \\ 0 \\ 3 \end{pmatrix}$

- **square matrix** $n \times n$ là ma trận có số cột và số hàng bằng $n$.

$\begin{pmatrix} 1 & 0 & 3 \\
 2 & -1 & 4 \\ 
 11 & 7 & 11 
 \end{pmatrix}$

- **zero matrix** là ma trận với tất cả các phần từ bằng $0$.

$\begin{pmatrix} 0 & 0 & 0 \\
 0 & 0 & 0 \\ 
 0 & 0 & 0 
 \end{pmatrix}$

 #### Ma trận đường chéo (diagonal matrices)

 Ma trận đường chéo là ma trận vuông mà tất cả các **leading entry** (phần tử non-zero đầu tiên của hàng) nằm trên đường chéo chính **leading (main) diagonal** (từ đỉnh trái đến đáy phải, không cần thiết phải là ma trận vuông). 

$\begin{pmatrix} \color{red}{1} & 0 & 0 \\
 0 & \color{red}{3} & 0 \\ 
 0 & 0 & \color{red}{7} 
 \end{pmatrix}$
 
 Ma trận số **scalar matrix** là ma trận tam giác mà trên đường chéo chính chứa cùng một giá trị scalar.

 $\begin{pmatrix} \color{red}{7} & 0 & 0 \\
 0 & \color{red}{7} & 0 \\ 
 0 & 0 & \color{red}{7} 
 \end{pmatrix}$

 Đường chéo phụ (**secondary diagonal**):

  $\begin{pmatrix} 0 & 0 & \color{red}{6} \\
 0 & \color{red}{4} & 0 \\ 
 \color{red}{-3} & 0 & 0 
 \end{pmatrix}$


Một **ma trận đơn vị** (identity matrix or unit matrix) kích cỡ $n$ là ma trận vuông $n \times n$ với tất cả các phần tử trên đường chéo chính bằng $1$, các phần từ còn lại bằng $0$. Ký hiệu $I_n$ hay $I$.

Ví dụ:

 $I_{3,3} = \begin{pmatrix} 1 & 0 & 0 \\
 0 &1 & 0 \\ 
 0 & 0 & 1
 \end{pmatrix}$

**Trace of matrix** là tổng các giá trị phần tử trên đường chéo chính. Kí hiệu $tr(A) = a_{1,1} + a_{2,2} + ... + a_{n,n}$

#### Ma trận tam giác

Ma trận tam giác trên (**upper traingular matrix**): $a_{i.j} = 0, if i > j$.

 $\begin{pmatrix} 7 & 2 & 1 \\
 \color{red}{0} & 7 & 4 \\ 
 \color{red}{0} & \color{red}{0} & 7
 \end{pmatrix}$

 Ma trận tam giác dưới (**lower triangular matrix**): $a_{i.j} = 0, if i < j$.

 $\begin{pmatrix} 7 & 0 & 0 \\
3 & 7 & \color{red}{0} \\ 
 1 & 2 & 2
 \end{pmatrix}$

Ma trận đối xứng (**symmetric matrix**) là ma trận vuông khi $a_{i,j} = a_{j,i}$

$\begin{pmatrix}
7 & \color{red}{0} & \color{green}{1} \\
\color{red}{0} & 7 & \color{blue}{2}  \\
\color{green}{1} & \color{blue}{2} & 2
\end{pmatrix}$



### Operations with matrices

Tập trung vào các phép toán: Nhân với scalar, cộng trừ ma trận và chuyển vị ma trận.

#### Scalar multiplication

Ta có thể nhân ma trận bởi bất kỳ số nào (**scalar**) - Bởi vì số này scale ma trận. 

$
-2 \times \begin{pmatrix}
-2 & 0 & -3 \\
1 & -1 & 0
\end{pmatrix} = \begin{pmatrix}
\color{red}{-2} \times -2 & \color{red}{-2} \times 0 & \color{red}{-2} \times -3 \\
\color{red}{-2} \times 1 & \color{red}{-2} \times -1 & \color{red}{-2} \times 0
\end{pmatrix} =
\begin{pmatrix}
4 & 0 & 6 \\
-2 & 2 & 0
\end{pmatrix}$

__Notes__:
*Ta dùng scalar multiplication để đơn giản ma trận cho việc giải hệ phương trình (system of equation). Chọn một constant để đơn giản ma trận.*

#### Matrix sum, difference

Muốn cộng, trừ ma trận thì hai ma trận phải cùng size, **cùng số chiều**.

$
\begin{pmatrix}
3 & 0 & -3 \\
2 & -1 & 8
\end{pmatrix} + \begin{pmatrix}
1 & 2 & 3 \\
3 & 1 & -7
\end{pmatrix} =
\begin{pmatrix}
3+1 & 0+2 & -3+3 \\
2+3 & -1+1 & 8-7
\end{pmatrix} = 
\begin{pmatrix}
4 & 2 & 0 \\
5 & 0 & 1
\end{pmatrix}
$

$
\begin{pmatrix}
1 & 2 \\
2 & 3 
\end{pmatrix} - \begin{pmatrix}
-2 & 0 \\
2 & 5 
\end{pmatrix} =
\begin{pmatrix}
1 - (-2) & 2 - 0 \\
2 - 2 & 3 - 5
\end{pmatrix} = 
\begin{pmatrix}
3 & 2 \\
0 & -2 
\end{pmatrix}
$

Tính chất: 

- $-A = (-1) \times A$

- $A - B = A + (-B)$


#### Transposed matrix


Từ ma trận $A$, ma trận chuyển vị $A^{T}$ có được bằng hoán đổi cột với dòng.

Ví dụ:

$
F_{2,3} = \begin{pmatrix}
1 & 2 \\
1 & 2  \\
1 & 2
\end{pmatrix} 
$

Tìm ma trận transposed matrix:

$
F^{T}_{2,3} = \begin{pmatrix}
1 & 2 \\
1 & 2  \\
1 & 2
\end{pmatrix}^{T} =  \begin{pmatrix}
1 & 1 & 1 \\
2 & 2 & 2  \end{pmatrix}
$

Chuyển vị có thể hình dung bằng việc reflect các phần tử qua đường chéo chính.





### Biểu diễn ma trận của hệ phương trình tuyến tính
(Matrix representation of systems of linear equations)


#### Hệ phương trình tuyến tính
(Systems of linear equations)

Nhìn nhanh qua hệ phương trình tuyến tính mà ai cũng đã học:

$
  \begin{cases}
   x + y = 1\\
    2x - y = 3
  \end{cases}
$

Với biến unknown là $x, y$, số đứng trước là **constant terms**.

#### Biến đổi một hệ phương trình tuyến tính sang dạng ma trận

Phần này đi nhanh, cũng đã quen thuộc. Ví dụ:

$
  \begin{cases}
   x + 2y + 3z = 4\\
    2x + 3y - z = 2 \\
    7x - 3y + 2z = 1 
  \end{cases}
$

Dạng ma trận sẽ là:

$
\left(
    \begin{array}{ccc|c}
     1 & 2 & 3 & 4 \\
     2 & 3 & -1 & 2 \\
     7 & -3 & 2 & 1
    \end{array}\
\right)
$

Ma trận như này được gọi là **extended**. Bên trái đường thẳng phân chia được gọi là **matrix of the unknowns**.
